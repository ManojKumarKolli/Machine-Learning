```python
X = df.drop(columns=['price'])
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_model.score(X_train, y_train)
```

```python
lr_model.score(X_test, y_test)
```

```python
# removing the outliers
bedrooms_max_thresould = df['bedrooms'].quantile(0.9999)
bedrooms_min_thresould = df['bedrooms'].quantile(0.001)
```

```python
df = df[(df['bedrooms'] < bedrooms_max_thresould) & (df['bedrooms'] > bedrooms_min_thresould)]
```

```python
df.describe()
```

```python
sqft_max_thresould = df['sqft_living'].quantile(0.9995)
sqft_min_thresould = df['sqft_living'].quantile(0.0005)

sqft_max_thresould
```

```python
df = df[df['sqft_living'] < sqft_max_thresould]
```

```python
df.describe()
```

```python
X = df.drop(columns=['price'])
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_out_model = LinearRegression()
lr_out_model.fit(X_train, y_train)
lr_out_model.score(X_test, y_test)
```
