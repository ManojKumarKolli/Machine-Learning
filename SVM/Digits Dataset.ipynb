```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
```

```python
df
```

```python
target = pd.DataFrame(digits.target, columns=['digit'])
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.20, random_state=42)
```

```python
digits_model = SVC(C = 100,kernel='linear')
```

```python
digits_model.fit(X_train, y_train.values.ravel())
```

```python
digits_model.score(X_test, y_test)
```
